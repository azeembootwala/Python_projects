```python
import os,pandas
from geopy.geocoders import Nominatim
nom=Nominatim()
os.listdir()
```

```python
df1=pandas.read_json("supermarkets.json")
df1
```

```python
df1["Address"]=df1["Address"]+", "+df1["City"]+", "+df1["State"]
df1
```

```python
df1.ix[df1.index[4],df1.columns[0]]
```

```python
df1["Latitude"]=df1.shape[0]*""
df1["Longitude"]=df1.shape[0]*""
df1
```

```python
for i in range(0,6):
    result=nom.geocode(df1.ix[df1.index[i],df1.columns[0]])
    if result!=None:
        df1.ix[df1.index[i],"Latitude"]=result.latitude
        df1.ix[df1.index[i],"Longitude"]=result.longitude
    else:
        None
        
df1
    
    
    
    
    
```

```python
df1["Codinates"]=df1["Address"].apply(nom.geocode)
df1["Latitude"]=df1["Codinates"].apply(lambda x :x.latitude if x!=None else None)
df1["Longitude"]=df1["Codinates"].apply(lambda x :x.longitude if x!=None else None)
df1
```
